# Sector Feature Summary

- Use this notebook to inspect the stock-level and sector-level feature-search rollups.
- The source CSVs are generated by `experiments/search/summarize_vn_sector_features.py`.


```python
from __future__ import annotations

from pathlib import Path

import pandas as pd

ROOT = Path.cwd()
TRAINING_RUNS_DIR = ROOT / "data" / "processed" / "assets" / "data_info_vn" / "history" / "training_runs"
STOCK_SUMMARY_CSV = TRAINING_RUNS_DIR / "stock_sector_search_summary.csv"
SECTOR_SUMMARY_CSV = TRAINING_RUNS_DIR / "sector_feature_final_summary.csv"
FEATURE_LONG_CSV = TRAINING_RUNS_DIR / "sector_feature_long.csv"
STOCK_SUMMARY_CSV, SECTOR_SUMMARY_CSV
```

```python
stock_summary_df = pd.read_csv(STOCK_SUMMARY_CSV) if STOCK_SUMMARY_CSV.exists() else pd.DataFrame()
sector_summary_df = pd.read_csv(SECTOR_SUMMARY_CSV) if SECTOR_SUMMARY_CSV.exists() else pd.DataFrame()
feature_long_df = pd.read_csv(FEATURE_LONG_CSV) if FEATURE_LONG_CSV.exists() else pd.DataFrame()

print(f"stock summary rows: {len(stock_summary_df)}")
print(f"sector summary rows: {len(sector_summary_df)}")
sector_summary_df.head(20)
```

```python
if not feature_long_df.empty:
    display(
        feature_long_df.groupby(["sector", "feature"], as_index=False)
        .agg(stocks_hit=("stock", "nunique"), appearances=("feature", "size"), mean_test_rel_score=("test_rel_score", "mean"))
        .sort_values(["stocks_hit", "appearances", "mean_test_rel_score"], ascending=[False, False, False])
        .head(25)
    )
else:
    print("Run `python3 experiments/search/summarize_vn_sector_features.py` to regenerate the rollup CSVs.")
```
