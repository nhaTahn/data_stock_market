```python
from __future__ import annotations

from pathlib import Path

import pandas as pd

ROOT = Path.cwd()
RUN_BASE = ROOT / "data" / "processed" / "assets" / "data_info_vn" / "history" / "training_runs"
DETAILS_CSV = RUN_BASE / "reports" / "target_mode_comparison_details.csv"
SUMMARY_CSV = RUN_BASE / "reports" / "target_mode_comparison_summary.csv"
DETAILS_CSV, SUMMARY_CSV
```

```python
details_df = pd.read_csv(DETAILS_CSV) if DETAILS_CSV.exists() else pd.DataFrame()
summary_df = pd.read_csv(SUMMARY_CSV) if SUMMARY_CSV.exists() else pd.DataFrame()

print(f"details rows: {len(details_df)}")
print(f"summary rows: {len(summary_df)}")
summary_df
```

```python
if not details_df.empty:
    display(
        details_df.sort_values(
            ["target_mode", "best_lstm_test_rel_score", "best_lstm_val_rel_score"],
            ascending=[True, False, False],
        ).head(20)
    )
else:
    print("Run `python3 experiments/analysis/compare_target_modes.py` to generate the CSV artifacts.")
```
